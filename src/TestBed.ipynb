Run every method in the RunManager, and examine the output to make sure it looks more or less reasonable.

```python
import networkx as nx
import matplotlib.pyplot as plt

from TestBed import RandomAttacker, RandomDefender
from RunManager import RunManager
```

```python
attacker = RandomAttacker()
defender = RandomDefender()
```

Now create the actual RunManager. I start with 20 nodes and 45 edges, mostly because it seemed to make the visualization look nice at the moment.

```python
manager = RunManager(attacker, defender, 2, nx.density, 20, 45)
```

Building the network. The current RandomDefender verbosely lists out each edge it creates, for verification purposes. Obviously, we'll remove that soon, or at least make it optional.

```python
manager.build_initial_network()
# Get the positions once, to keep the network static in the future.
positions = nx.spring_layout(manager.G) 
nx.draw_networkx(manager.G, positions)
```

Now we subject the network to attack, see which node gets removed (the attacker also prints out her decision here) and which nodes are now disconnected.

```python
manager.attack_network()
print manager.nodes_lost_edges
nx.draw_networkx(manager.G, positions)
```

To end the round, we rewire the network, again displaying the results.

```python
manager.rewire_network()
nx.draw_networkx(manager.G, positions)
```
